# Task 1 — 2-Qubit Circuit with 90° Phase Shift

## What is happening?

- `H gate` creates superposition (qubit becomes both 0 and 1)
- `S gate` applies a 90° phase shift
- `CX gate` entangles two qubits together

The Q-sphere helps visualize the quantum state.


````python
import sys
!{sys.executable} -m pip install qiskit pylatexenc

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere
import matplotlib.pyplot as plt


````

````python

# Create 2-qubit circuit
qc = QuantumCircuit(2)

# Put qubit into superposition
qc.h(0)

# Apply 90° phase shift
qc.s(0)

# Entangle qubits
qc.cx(0, 1)

# Draw circuit
qc.draw('mpl')
````

````python
# Generate statevector
state = Statevector.from_instruction(qc)

# Plot Q-sphere
plot_state_qsphere(state)
````


# Task 2 — QASM Circuit Import and Execution

## What is QASM?

QASM is a text-based language for describing quantum circuits.

Example:
```qasm
h q[0];
cx q[0], q[1];
```

This creates superposition and entanglement.


````python
import sys
!{sys.executable} -m pip install qiskit-aer

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile

# Define QASM code
qasm_code = '''
OPENQASM 2.0;
include "qelib1.inc";

qreg q[2];
creg c[2];

h q[0];
cx q[0], q[1];
measure q -> c;
'''

# Import circuit from QASM
qc2 = QuantumCircuit.from_qasm_str(qasm_code)

# Draw circuit
qc2.draw('mpl')
````

````python

# Simulate the circuit
simulator = AerSimulator()

compiled = transpile(qc2, simulator)
job = simulator.run(compiled, shots=1000)

result = job.result()
counts = result.get_counts()

print("Measurement Results:")
print(counts)
````


# Task 3 — Random Quantum Circuit

## Goal

Create a random quantum circuit manually without using auto dump methods.

This helps understand how different quantum gates affect qubits.


````python

from qiskit import QuantumCircuit
import random

qc3 = QuantumCircuit(3)

gates = ['h', 'x', 'y', 'z']

# Apply random gates
for i in range(5):
    gate = random.choice(gates)
    qubit = random.randint(0, 2)

    if gate == 'h':
        qc3.h(qubit)
    elif gate == 'x':
        qc3.x(qubit)
    elif gate == 'y':
        qc3.y(qubit)
    elif gate == 'z':
        qc3.z(qubit)

# Add entanglement
qc3.cx(0, 1)
qc3.cx(1, 2)

# Draw circuit
qc3.draw('mpl')
````
